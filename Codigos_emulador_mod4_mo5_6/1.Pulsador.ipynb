## 2. Declarar puerto

Declarar el modo del GPIO

Declarar un pin como entrada

```python
GPIO.setmode(GPIO.BCM)
GPIO.setup(18, GPIO.IN, pull_up_down=GPIO.PUD_UP)
```

## 3. Creacion del bucle

Sentecia infinita "while True:".

Crear variable de entrada.

```python
while True:
    input_state = GPIO.input(18)
```

## 4. Condicional

Condicionamos la entrada para la accion "print".

```python
if input_state == False:
        print('Button Pressed')
        time.sleep(0.2)
```

## Codigo completo

```python
#!/usr/bin/python3
__author__ = "Cristian Moyano"
__copyright__ = "Cristian Moyano, 2020"
__credits__ = ["Alexander Saravia", "Cristian Moyano"]
__license__ = "CC BY-NC-SA 4.0"
__version__ = "1.0.0"
__maintainer__ = "Cristian Moyano"
__email__ = "cristian.moyano@ieee.org"
__status__ = "Development"

from EmulatorGUI import GPIO
#import RPi.GPIO as GPIO
import time
GPIO.setmode(GPIO.BCM)
GPIO.setup(18, GPIO.IN, pull_up_down=GPIO.PUD_UP)
while True:
    input_state = GPIO.input(18)
    if input_state == False:
        print('Button Pressed')
        time.sleep(0.2)
```
